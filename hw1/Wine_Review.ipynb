### 通过属性的相关关系来填补缺失值
> ***抽象为回归问题，拟利用多项式回归进行拟合，拟合结果为价格***

```python
import sklearn.pipeline as pl
import sklearn.preprocessing as sp
import sklearn.linear_model as lm
```

```python
def fillNanWithPolyRegress(df, title):
    # 把已有的数值型特征取出来输入到RandomForestRegressor中
    new_df = copy.deepcopy(df)
    df_numberic = df[['price', 'points']]
    # 已知该特征和未知该特征
    training_data = df_numberic[df_numberic.price.notnull()].values
    test_data = df_numberic[df_numberic.price.isnull()].values
    
    tr_X = training_data[:, 1].reshape(-1, 1)
    tr_y = training_data[:, 0].reshape(-1, 1)
    
    model = pl.make_pipeline(sp.PolynomialFeatures(10), lm.LinearRegression())
    
    model.fit(tr_X,tr_y)
    # 用得到的模型进行未知特征值预测
    predicted = model.predict(test_data[:, 1].reshape(-1, 1))
    # 用得到的预测结果填补原缺失数据
    num_idx = new_df.price.isnull()
    new_df.loc[num_idx, 'price'] = predicted
    
    #绘制缺失值的散点图
    new_df.loc[num_idx].plot.scatter(x='points', y='price',title=title)
    return new_df
```

```python
# 计算相关系数
print('wine_150k的person相关系数矩阵：')
display(wine_data_150k.corr())
print('wine_130k的person相关系数矩阵：')
display(wine_data_130k.corr())
# 绘制总体散点图
wine_data_150k.plot.scatter(title='wine_150k', x='points', y='price')
wine_data_130k.plot.scatter(title='wine_130k', x='points', y='price')
```

> ***综上，可以发现分数和价格之间呈现正相关的关系***

```python
# 利用多项式拟合训练数据后填补缺失的价格
print('拟合结果为:')
wine_data_150k_3 = fillNanWithPolyRegress(wine_data_150k, title='wine_150k')
wine_data_130k_3 = fillNanWithPolyRegress(wine_data_130k, title='wine_130k')
```

> ***抽象为分类问题，拟利用随机森林进行分类，类别即为价格，特征选取国家、分数、品种和酒庄***

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
```

```python
def fillNanWithRFC(df, title):
    new_df = copy.deepcopy(df)
    df_data = df[[ 'price', 'points', 'country', 'variety', 'winery']]
    trans_data = df_data.values
    training_data = df_data[df_data.price.notnull()].values
    test_data = df_data[df_data.price.isnull()].values
    
    encoder_list = [LabelEncoder() for i in range(3)]
    
    for i in range(3):
        encoder_list[i].fit(trans_data[:,i + 2])
        training_data[:, i + 2] = encoder_list[i].transform(training_data[:,i + 2])
        test_data[:, i + 2] = encoder_list[i].transform(test_data[:, i + 2])
        
    tr_X = training_data[:, 1:].astype(np.float64)
    tr_y = training_data[:, 0].reshape(-1, 1).astype(np.float64)
    
    rfc = RandomForestClassifier(random_state=0, n_estimators=20,  n_jobs=-1)
    rfc.fit(tr_X,tr_y)
    # 用得到的模型进行未知特征值预测
    predicted = rfc.predict(test_data[:, 1:].astype(np.float64))
#     用得到的预测结果填补原缺失数据
    num_idx = new_df.price.isnull()
    new_df.loc[num_idx, 'price'] = predicted
    
    #绘制缺失值的散点图
    new_df.loc[num_idx].plot.scatter(x='points', y='price',title=title)
    return new_df
```
