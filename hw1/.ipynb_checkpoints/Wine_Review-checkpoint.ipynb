## 数据集：Wine Reviews
### 数据描述:
>**id:** 编号  
 **country:** 葡萄酒来源的国家  
 **description:** 葡萄酒描述  
 **designation:** 葡萄园名称  
 **points:** 葡萄酒评分  
 **price:** 葡萄酒价格  
 **province:** 葡萄酒来源的省或州  
 **region_1:** 葡萄酒产区  
 **region_2:** 葡萄酒具体产区  
 **variety:** 葡萄的品种  
>**winery:** 酿酒厂 
----
>**taster_name:** 品鉴师  
 **taster_twitter_handle:** 品鉴师推特  
>**title:** 品鉴师头衔

```python
file_path = '/Users/Jexer/Documents/Dataset/Wine_Reviews'
file_name_1 = 'winemag-data_first150k.csv'
file_name_2 = 'winemag-data-130k-v2.csv'

wine_data_150k = loadData(file_path, file_name_1).rename(columns={'Unnamed: 0':'id'})
wine_data_130k = loadData(file_path, file_name_2).rename(columns={'Unnamed: 0':'id'})

display(wine_data_150k.head(5))
display(wine_data_130k.head(5))
```

```python
print('len of wine_150k:', len(wine_data_150k))
for column in wine_data_150k.columns:
    print(column, wine_data_150k[column].isnull().any(), wine_data_150k[column].nunique())
print('-' * 40)
print('len of wine_130k:', len(wine_data_130k))
for column in wine_data_130k.columns:
    print(column, wine_data_130k[column].isnull().any(), wine_data_130k[column].nunique())
```

```python
for column in wine_data_150k.columns:
    idx = wine_data_150k[column].isnull()
    if idx.any():
        print('-' * 50, column, '-' * 50)
        display(wine_data_150k[idx].head(3))
```
